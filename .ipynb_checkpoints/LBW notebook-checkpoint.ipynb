```python
data=df[['v012','v201','v212','v218','v445','v446','v155','v190','v729','b5','m18']]
print(data.head())
```

```python
data_n=data.dropna()
print(data_n.head())
```

```python
data_n.describe()
```

```python
h=data_n.groupby(['v729'])
print(h.describe())
```

```python
def education(x):
    if x=='no education':
        return 1
    if x=='incomplete primary':
        return 1
    if x=='complete primary':
        return 2
    if x=='incomplete secondary':
        return 2
    if x=='complete secondary':
        return 3
    if x=='higher':
        return 4
data_n['v729']=data_n['v729'].apply(education)
data_n
```

```python
h=data_n.groupby(['v190'])
print(h.describe())
```

```python
def wealth(x):
    if x=='middle':
        return 2
    if x=='poorer':
        return 1
    if x=='poorest':
        return 1
    if x=='richer':
        return 3
    if x=='richest':
        return 3
data_n['v190']=data_n['v190'].apply(wealth)
data_n
```

```python
h=data_n.groupby(['v155'])
print(h.describe())
```

```python
def edu_w(x):
    if x=='able to read only parts of sentence':
        return 2
    if x=='able to read whole sentence':
        return 3
    if x=='cannot read at all':
        return 1
    
data_n['v155']=data_n['v155'].apply(edu_w)
data_n
```

```python
h=data_n.groupby(['m18'])
print(h.describe())
```

```python
def lbw(x):
    if x=='average':
        return 1
    if x=='larger than average':
        return 1
    if x=='very large':
        return 1
    if x=='smaller than average':
        return 2
    if x=='very small':
        return 2
    
data_n['m18']=data_n['m18'].apply(lbw)
data_n
```

```python
final=data_n.drop(columns=['edu','b5'])
final
```

```python
final.to_csv('/Users/pcworld/Desktop/419/final_revised.csv',index=False)
```
